## Importing the libraries

```
import ocrmypdf
import pdfplumber
```

## Opening the pdf and output out of it

```
work_sample = ('sample for detection.pdf')
```

```
with pdfplumber.open(work_sample) as pdf:
  page = pdf.pages[0]
  text = page.extract_text()
  print(text)
```
